```python
def extended_euclidean(a, b):
    r1 = a
    r2 = b
    s1 = int(1)
    s2 = int(0)
    t1 = int(0)
    t2 = int(1)

    while r2 > 0:
        q = r1//r2
        r = r1-q * r2
        r1 = r2
        r2 = r
        
        s = s1-q * s2
        s1 = s2
        s2 = s
        
        t = t1-q * t2
        t1 = t2
        t2 = t

    if t1 < 0:
        t1 = t1 % a

    return (r1, t1)
```

```python
def euclid(a, b):
    if a == 0:
        return b
    
    return euclid(b%a, a)
```

```python
p = int(input("Enter the value of p:")) #p = 823
q = int(input("Enter the value of q:")) #q = 953

n = p*q
phiN = (p-1)*(q-1)
```

```python
# Generate Encryption Key
# in range 1<e<Pn
key = []
for i in range(2, phiN):

    gcd = euclid(phiN, i)

    if gcd == 1:
        key.append(i)
```

```python
# Select an encryption key from the above list
e = int(355) # Public key encryption
print("Public Key Encryption: {E} {N}".format(E = e, N = n) )
```

```python
# Obtain inverse of encryption key.
r, d = extended_euclidean(phiN, e)
if r == 1:
    d = int(d) # Calculate Private Key
    print("Decryption/Private key {D} {N}: ".format(D = d, N = n))

else:
    print("Multiplicative inverse for the given encryption key does not exist. Choose a different encryption key ")
```

```python
# Signature is created by Alice
S = (M**d) % n
```

```python





# Alice sends M and S both to Bob
# Bob generates message M1 using the
# signature S, Alice's public key e
# and product n.
M1 = ((S)**e) % n

# If M = M1 only then Bob accepts
# the message sent by Alice.

if M == M1:
    print("As M = M1, Accept the message sent by Alice")
else:
    print("As M not equal to M1,Do not accept the message sent by Alice ")
```
